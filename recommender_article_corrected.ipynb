```python
all_data = pd.merge(sessions, news, on =['article_id'])
```

```python
all_data = all_data.iloc[:,1:].copy()
```

```python
all_data['count'] = 1
```

```python
matrix = all_data[['user_id', 'article_id', 'count']].pivot_table(
    index=['user_id'],
    columns=['article_id'],
    values=['count'],
    aggfunc='sum'
)['count']
```

```python
matrix.fillna(0, inplace=True)
```

```python
matrix
```

```python
def get_similar_articles(article_id, n_recommendations=5, same_category=False):
    similar = matrix.corrwith(matrix[article_id]).sort_values(ascending=False)
    corr_similar = pd.DataFrame(similar, columns=['correlation'])
    
    corr_with_news = pd.merge(
        left=corr_similar, 
        right=news, 
        on='article_id')
    
    
    if same_category: 
        article_category = news[news['article_id']==article_id]['category'].iloc[0]
        corr_with_news = corr_with_news[corr_with_news['category']==article_category]
    
    return corr_with_news.head(n_recommendations)
```

```python
get_similar_articles('content_250', n_recommendations=5, same_category=True)
```
